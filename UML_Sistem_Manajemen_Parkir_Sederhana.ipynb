```python
import subprocess
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#@title Class Diagram
sequence_diagram_code = """
@startuml
class ParkingSlot {
    +int slotID
    +boolean isAvailable
    +void checkAvailability()
    +void markAsOccupied(String licensePlate)
    +void markAsFree()
}

class Vehicle {
    +String licensePlate
    +String entryTime
    +String exitTime
    +void setEntryTime()
    +void setExitTime()
    +int calculateParkingDuration()
}

class ParkingHistory {
    +int historyID
    +String licensePlate
    +String slotID
    +String entryTime
    +String exitTime
    +void recordEntry(String licensePlate, int slotID, String entryTime)
    +void recordExit(String licensePlate, int slotID, String exitTime)
}

class Database {
    +void connect()
    +void insertParkingSlot(ParkingSlot slot)
    +void insertHistoryRecord(ParkingHistory record)
    +List<ParkingSlot> getAvailableSlots()
    +List<ParkingHistory> getParkingHistory()
}

ParkingSlot "1" -- "*" Vehicle
ParkingHistory "1" -- "*" Vehicle
Database "1" -- "*" ParkingSlot
Database "1" -- "*" ParkingHistory
@enduml
"""

# Menyimpan kode di file .puml
with open("class_diagram.puml", "w") as file:
    file.write(sequence_diagram_code)

# Menjalankan PlantUML untuk mengubah .puml menjadi gambar PNG
subprocess.run(["plantuml", "class_diagram.puml"])

# Menampilkan gambar dengan matplotlib
img = mpimg.imread("class_diagram.png")
plt.imshow(img)
plt.axis('off')  # Menyembunyikan sumbu
plt.show()
```

```python
import subprocess
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#@title Class Diagram
sequence_diagram_code = """
@startuml
actor "User" as user
actor "Admin" as admin

usecase "Check Available Parking Slots" as checkSlots
usecase "Park Vehicle" as parkVehicle
usecase "Retrieve Parking History" as retrieveHistory
usecase "Update Slot Availability" as updateSlot
usecase "Monitor Vehicle Entry/Exit" as monitorVehicle

user --> checkSlots
user --> parkVehicle
user --> retrieveHistory

admin --> updateSlot
admin --> monitorVehicle

checkSlots --> parkVehicle : <<include>>
parkVehicle --> monitorVehicle : <<extend>>
retrieveHistory --> monitorVehicle : <<include>>
@enduml
"""

# Menyimpan kode di file .puml
with open("usecase_diagram.puml", "w") as file:
    file.write(sequence_diagram_code)

# Menjalankan PlantUML untuk mengubah .puml menjadi gambar PNG
subprocess.run(["plantuml", "usecase_diagram.puml"])

# Menampilkan gambar dengan matplotlib
img = mpimg.imread("usecase_diagram.png")
plt.imshow(img)
plt.axis('off')  # Menyembunyikan sumbu
plt.show()
```
